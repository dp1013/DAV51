```r
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```r
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### Matplotlib

```r
plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=30, color='skyblue')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```r
species_labels = ['Setosa', 'Versicolor', 'Virginica']
species_counts = [np.sum(y == i) for i in range(3)]
plt.bar(species_labels, species_counts, color=['red', 'green', 'blue'])
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar Chart of Iris Species')
plt.show()
```

```r
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', s=60)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.colorbar(label='Species')
plt.show()
```

```r
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_labels, autopct='%1.1f%%', startangle=140, colors=['red', 'green', 'blue'])
plt.title('Pie Chart of Iris Species Distribution')
plt.axis('equal')
plt.show()
```

```r
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', s=60)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.set_title('3D Scatterplot of Iris Data')
plt.show()
```

### Seaborn

```r
iris = sns.load_dataset('iris')
```

```r
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```r
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

```r
correlation_matrix = iris.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```r
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='scatter', hue='species')
plt.suptitle('Jointplot of Petal Length vs Petal Width', y=1.02)
plt.show()
```

```r
g = sns.FacetGrid(iris, col='species', hue='species')
g.map(sns.scatterplot, 'sepal_length', 'sepal_width')
g.add_legend()
plt.suptitle('FacetGrid of Sepal Length vs Sepal Width by Species', y=1.02)
plt.show()
```

### Plotly

```r
x = np.random.rand(100)
y = np.random.rand(100)
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers', marker=dict(color='blue')))
fig.update_layout(title='Interactive Scatter Plot', xaxis_title='X-axis', yaxis_title='Y-axis', hovermode='closest')
fig.show()
```

```r
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='sin(x)', line=dict(color='red')))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='cos(x)', line=dict(color='blue')))
fig.update_layout(title='Customized Line Plot', xaxis_title='X-axis', yaxis_title='Y-axis')
fig.add_annotation(text='Maximum', x=np.pi, y=1, showarrow=True)
fig.show()
```

```r
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(title='3D Surface Plot', scene=dict(xaxis_title='X-axis', yaxis_title='Y-axis', zaxis_title='Z-axis'))
fig.show()
```

### ggplot2

```r
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) +
  geom_point() +
  labs(title = "Scatterplot of Sepal Length vs Sepal Width with Species Encodings")
```

```r
ggplot(iris, aes(x = Species, y = Sepal.Length, fill = Species)) +
  geom_boxplot() +
  labs(title = "Boxplot of Sepal Length by Species")
```

```r
ggplot(iris, aes(x = Sepal.Width, fill = Species)) +
  geom_histogram(binwidth = 0.2, alpha = 0.7) +
  labs(title = "Histogram of Sepal Width by Species")
```

```r
ggplot(iris, aes(x = Sepal.Length, y = Petal.Length, color = Species)) +
  geom_line() +
  labs(title = "Line Plot of Sepal Length vs Petal Length with Species Encodings")
```

```r
ggplot(iris, aes(x = Species, fill = Species)) +
  geom_bar() +
  labs(title = "Bar Plot of Species Counts")
```

### plotly

```r
x <- 1:10
y <- x^2
xyplot(y ~ x, type = "l")
```

```r
x <- 1:10
y <- 1:10
z <- rnorm(100)
cloud(z ~ x*y, main = "3D Lattice Graph")
```

```r
x <- seq(-pi, pi, length.out = 100)
y <- seq(-pi, pi, length.out = 100)
z <- outer(x, y, function(x, y) sin(sqrt(x^2 + y^2)))
levelplot(z, xlab = "x", ylab = "y",
		main = "2D Sin Function")
```
