```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Event Handling in D3.js</title>
    <script src="https://d3js.org/d3.v7.min.js"></script>
    <style>
        .circle {
            fill: steelblue;
            stroke: white;
            stroke-width: 2px;
        }

        .circle:hover {
            fill: orange;
        }
    </style>
</head>
<body>
    <svg width="400" height="200"></svg>

    <script>
        const svg = d3.select("svg");

        // Data for circles
        const data = [
            { x: 50, y: 50 },
            { x: 150, y: 100 },
            { x: 250, y: 150 }
        ];

        // Draw circles
        const circles = svg.selectAll("circle")
            .data(data)
            .enter().append("circle")
            .attr("class", "circle")
            .attr("cx", d => d.x)
            .attr("cy", d => d.y)
            .attr("r", 20)
            .on("mouseover", handleMouseOver)
            .on("mouseout", handleMouseOut)
            .on("click", handleClick)
            .on("mousemove", handleMouseMove)
            .on("dblclick", handleDoubleClick);

        // Function to handle mouseover event
        function handleMouseOver(d, i) {
            d3.select(this).attr("r", 25); // Increase circle size
        }

        // Function to handle mouseout event
        function handleMouseOut(d, i) {
            d3.select(this).attr("r", 20); // Restore circle size
        }

        // Function to handle click event
        function handleClick(d, i) {
            alert(`Clicked circle ${i + 1}`);
        }

        // Function to handle mousemove event
        function handleMouseMove(d, i) {
            const [x, y] = d3.pointer(event);
            console.log(`Mouse position: (${x}, ${y})`);
        }

        // Function to handle double click event
        function handleDoubleClick(d, i) {
            d3.select(this).attr("fill", "red"); // Change circle color
        }
    </script>
</body>
</html>
```
